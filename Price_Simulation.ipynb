Here we will see a simple example of how to simulate stock prices, assuming that the log return comes from the normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#number of time steps
T = 1000
```

```python
#initial price
P_0 = 10
```

```python
#drift
mu = 0.001
```

```python
# last log price
last_p = np.log(P_0)

log_returns = np.zeros(T)
prices = np.zeros(T)

for t in range(T):
  # sample a log return
  r = 0.01 * np.random.randn()

  # compute the new log price
  p = last_p + mu + r

  # store the return and price
  log_returns[t] = r + mu
  prices[t] = np.exp(p)

  # assign last_p
  last_p = p
```

```python
plt.figure(figsize = (20, 8))
plt.plot(prices)
plt.title("Simulated Prices");
```
